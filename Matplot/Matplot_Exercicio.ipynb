```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mês
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])


```

 Gráfico 1 - Total de produtos Vendidos por mês - Linha

```python
total = np.array([[ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ], 
                  [ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ],
                  [ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ],
                  [ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ],
                  [ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ],
                  [ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ]])

fig, all = plt.subplots()
fig.set_size_inches(10, 5.4)

totalSum = np.sum(total, axis = 0)


all.plot(mes, totalSum)
all.set_title("Total de produtos vendidos por mês\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
all.set_xlabel("Meses", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
all.set_ylabel("Vendas", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
```

Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha

```python
fig, prod = plt.subplots()
fig.set_size_inches(10, 5.4)


prod.plot(mes, creme_facial, label='Creme facial')
prod.plot(mes, limpeza_facial, label='Limpeza facial')
prod.plot(mes, pasta_dentaria, label='Creme dental')
prod.plot(mes, sabonete, label='Sabonete')
prod.plot(mes, shampoo, label='Shampoo')
prod.plot(mes, hidratante, label='Hidratante')

prod.set_title("Produtos Vendidos por mês\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
prod.set_xlabel("Meses", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
prod.set_ylabel("Produtos", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

prod.legend()
```

Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras

```python
fig, compare = plt.subplots()

fig.set_size_inches(10, 5.4)

compare.bar(mes, creme_facial, width=0.9, edgecolor="white", linewidth=0.7, label='Creme facial')
compare.bar(mes, limpeza_facial, width=0.9, edgecolor="white", linewidth=0.7, label='Limpeza facial')

compare.set_title("Comparativo ente creme e limpeza facial\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
compare.set_xlabel("Meses", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
compare.set_ylabel("Creme e limpeza facial", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

compare.legend()
plt.show()
```

 Gráfico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos (1000-1999, 2000-2999, ...)

```python
fig, histogram = plt.subplots()
fig.set_size_inches(20, 8)
faixas = np.array([4000, 8000, 12000, 16000, 20000])

histogram.hist(total, bins = faixas, edgecolor ='white')

histogram.grid()
plt.show()
```

 Gráfico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto

```python
fig, pizza = plt.subplots()

fig.set_size_inches(10, 5.4)

creme_facial = np.sum(creme_facial, axis = 0)
limpeza_facial = np.sum(limpeza_facial, axis = 0)
pasta_dentaria = np.sum(pasta_dentaria, axis = 0)
sabonete = np.sum(sabonete, axis = 0)
shampoo = np.sum(shampoo, axis = 0)
hidratante = np.sum(hidratante, axis = 0)


totalPizza = np.sum(total)
percentuais = np.array([creme_facial/totalPizza, limpeza_facial/totalPizza, pasta_dentaria/totalPizza, sabonete/totalPizza, shampoo/totalPizza, hidratante/totalPizza])
produtos = np.array(["Creme facial", "Limpeza facial", "Pasta de dente", "Sabonete", "Shampoo", "Hidratante"])

explodir = [0, 0, 0, 0.2, 0, 0]
pizza.pie(percentuais, labels= produtos, explode = explodir, shadow = True, autopct='%1.1f%%')

```
